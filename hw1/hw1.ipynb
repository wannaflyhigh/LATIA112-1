```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

df = pd.read_csv('111_land.csv', encoding='utf-8')
df
```

### 1.學生宿舍床位前十名的大學

```python
# 使用groupby合并相同学校名字的数据，然后对学生宿舍床位数求和
merged_df = df.groupby('學校名稱').sum().reset_index()

# 根据合并后的数据，根据学生宿舍床位數_床列进行降序排序
df_sorted = merged_df.sort_values(by='學生宿舍床位數_床', ascending=False)

# 获取拥有最多学生宿舍床位的学校
top_school = df_sorted.iloc[0]

# 打印该学校信息
print("學校名稱:", top_school['學校名稱'])
print("學生宿舍床位數:", top_school['學生宿舍床位數_床'])

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(df_sorted['學校名稱'][:10], df_sorted['學生宿舍床位數_床'][:10])
plt.xlabel('學校名稱')
plt.ylabel('學生宿舍床位數')
plt.title('學校擁有最多學生宿舍床位的前十所學校')
plt.xticks(rotation=45)
plt.tight_layout()

# 显示图表
plt.show()
```

### 6.各縣市大學數量

```python
# 使用groupby按縣市別合并数据，然后使用count方法计算每个縣市的大學数量
city_university_count = df.groupby('縣市別').count()['學校名稱'].reset_index()

# 为了更好的可视化，你可以按大學数量降序排序
city_university_count_sorted = city_university_count.sort_values(by='學校名稱', ascending=False)

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(city_university_count_sorted['縣市別'], city_university_count_sorted['學校名稱'])
plt.xlabel('縣市')
plt.ylabel('大學數量')
plt.title('各縣市大學数量統計')
plt.xticks(rotation=45)
plt.tight_layout()

# 显示图表
plt.show()
```

### 7.各縣市國立大學和私立大學的比值

```python
# 使用groupby按縣市別合并数据，然后使用value_counts方法计算每个縣市的国立和私立大学数量
city_university_counts = df['縣市別'].value_counts().reset_index()
city_university_counts.columns = ['縣市別', '大學數量']

# 创建一个新列，用于存储国立大学的数量
city_university_counts['國立大學數量'] = 0

# 使用循环遍历每个縣市，计算国立大学的数量
for index, row in city_university_counts.iterrows():
    city = row['縣市別']
    national_count = len(df[(df['縣市別'] == city) & (df['學校類型'] == '國立大學')])
    city_university_counts.loc[index, '國立大學數量'] = national_count

# 创建一个新列，用于存储私立大学的数量
city_university_counts['私立大學數量'] = city_university_counts['大學數量'] - city_university_counts['國立大學數量']

# 计算国立和私立大学的比值
city_university_counts['國立/私立比值'] = city_university_counts['國立大學數量'] / city_university_counts['私立大學數量']

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(city_university_counts['縣市別'], city_university_counts['國立/私立比值'])
plt.xlabel('縣市')
plt.ylabel('國立/私立比值')
plt.title('各縣市國立大學和私立大學的比值')
plt.xticks(rotation=45)
plt.tight_layout()

# 显示图表
plt.show()
```

### 8.哪個學校在臺北市有最多的校地面積

```python
# 選擇臺北市的學校
taipei_schools = df[df['縣市別'] == '臺北市']

# 計算每個學校的可使用校地的總面積
taipei_schools['總可使用校地面積'] = taipei_schools['可使用校地-校舍基地_平方公尺'] + taipei_schools['可使用校地-露天運動場地_平方公尺'] + taipei_schools['可使用校地-其他校地_平方公尺']

# 將學校按照總可使用校地面積降序排列，並選取前十行
top_10_schools = taipei_schools.sort_values(by='總可使用校地面積', ascending=False).head(10)

# 打印前十名學校的名稱和總可使用校地面積
print("在臺北市擁有最多可使用校地的前十名學校是:")
for index, row in top_10_schools.iterrows():
    print(row['學校名稱'], "-", row['總可使用校地面積'], "平方公尺")

# 繪製圖表
plt.figure(figsize=(10, 6))
plt.bar(top_10_schools['學校名稱'], top_10_schools['總可使用校地面積'])
plt.xlabel('學校名稱')
plt.ylabel('總可使用校地面積（平方公尺）')
plt.title('臺北市學校總可使用校地面積前十名')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
